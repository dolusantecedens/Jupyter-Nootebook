# Ćwiczenie 1

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
original_price = 120000
house_rate = 0.05
nperiods = 5
periods = np.arange(1,nperiods+1)
final_price = np.around(npf.fv(house_rate,nperiods,0,-original_price),2)
print(final_price)
```

Cena mieszkania po upływie 5 lat będzie wynosiła 153153.79 zł

# Ćwiczenie 2

```python
import numpy as np
import numpy_financial as npf
bank_rate = 0.12/12
bank_nperiods = 12 * 5
bank_periods = np.arange(0, bank_nperiods)
initial_investment=np.around(npf.pv(bank_rate,bank_nperiods,0, fv=-final_price),2)
print(initial_investment)
```

Aby zakupić dom po 5 latach należy wpłacić na lokatę w dniu dzisiejszym 84303.44 zł

# Ćwiczenie 3

```python
price_arr = -npf.fv(house_rate,periods,nperiods,original_price)
spread_price_arr = np.array([np.linspace(original_price,price_arr[0],13)[:-1],np.linspace(price_arr[0],price_arr[1],13)[:-1],
                            np.linspace(price_arr[1],price_arr[2],13)[:-1],np.linspace(price_arr[2],price_arr[3],13)[:-1],
                            np.linspace(price_arr[3],price_arr[4],13)[:-1]]).reshape(60)
bank_arr = -npf.fv(bank_rate, bank_periods,bank_nperiods,initial_investment)
plt.plot(spread_price_arr,label='cena mieszkania')
plt.plot(bank_arr,label='kwota na lokacie')
plt.legend()
plt.xlabel('Liczba miesięcy')
plt.ylabel('Wartość w zł')
```
